```python
import matplotlib.pyplot as plt
```

```python
# Set x-axis to an array of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Create line charts

```python
# Create a line plot using MATLAB method
# plot() --> gives line chart by default
plt.plot(x_axis,y_axis,label="Boston")

plt.title("Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.legend()

# to show the plot 
# plt.show()
```

```python
# Create a plot with OO interface method 
# Create a figure
fig, ax = plt.subplots()

# Define axes and make a plot by sending them as parameters
ax.plot(x_axis, y_axis)
```

```python
# another way of plotting using the above OO code as below: (Artist -> FigureBase -> Figure)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt() 
ax = plt.axes()
ax.plot(x_axis, y_axis, color='green', marker='D', linewidth=2, label="Boston")

# OO method: set attributes on ax object

# set y-axis limit
ax.set_ylim(0,45)

# set x axis using index
#ax.set_xlim(0,6)

# add x axis label
ax.set_xlabel('Date')

# add y axis label
ax.set_ylabel('Fare($)')

# set title for graph
ax.set_title('Fare by Month')

# add legend
ax.legend()

# add grid
ax.grid()
```

```python
# Create a line plot (MATLAB method).
# more code but readable compared to format strings
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# default line width = 1
# if you need to add another line to the same chart for the same x axis values
# plt.plot(x_axis, y_axis2)
# plt.plt(x_axis, y_axis3)

# can also write above with fmt (not very readable code) : marker,linestyle,color
# 'bo' --> blue oooooo for one line ,'r+' --> red ++++ for another line
# When conflicting with fmt, keyword arguments take precedence.
# default linestyle: solid
# color: hex values can be used
# print(plt.style.available) pick the style you want and use it by plt.style.use('ggplot')
# if using custom styles, you do not have to use fmt strings and other params in plot function

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend. labels created in plot used here
plt.legend()

#plt.tight_layout # to improve padding


# change figure size to width 20 and height 10 pixels
# plt.subplots(figsize=(20, 10))
plt.figure(figsize=(20, 10))

#plt.savefig(fare.png) # saves image to current directory
```

## Create bar graphs

```python
# Create a vertical bar graph using MATLAB methods
# bar() --> create a vertical bar 
#plt.barh() for horizontal bar.
plt.bar(x_axis,y_axis, color="green", label="Boston")

# you can also add plt.plot(x_axis, y_axis2) to give a line graph on top of bar graph for comparison
# to get a bar beside the original data, use numpy to set x indexes and shift index by bar width and set width
# width= 0.25 (default is 0.8)
# x_indexes = np.arange(len(x_axis))
# plt.bar(x_indexes-width, y_1, width=width)
# plt.bar(x_indexes,y_2,width=width)
# plt.bar(x_indexes+width,y_2,width=width)
# plt.xticks(ticks=x_indexes,labels=x_axis)

# Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Add a title and legend
plt.title("Fare by Month")
plt.legend()
```

```python
# Create a horizontal bar graph
plt.barh(x_axis,y_axis, color="m", label="Boston")

# The months are in wrong order. Jan should be the first month on y axis.
# To invert the y-axis to have the months in ascending order, use the gca() method.
# chain the gca() method to the invert_yaxis() 
plt.gca().invert_yaxis()
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Switch axes 
plt.barh(y_axis,x_axis)

# The below graph is not a good way to represent the data because the bar should convey an amount on the 
# x-axis—not a date. Trying to determine each month's average fare is more challenging in this chart.
```

```python
# Create bar chart with OO approach

# Create vertical bar with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="New York")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis, color="c", label="Chicago")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()

```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label="Chicago")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
#labels read top to bottom
ax.invert_yaxis()
```

```python
# MATLAB method: there are two ways to create scatter plots: using plt.plot() ans plt.scatter()
# Create a scatter plot with plot()
plt.plot(x_axis,y_axis,'o')
```

```python
# Create a scatter plot with scatter()
plt.scatter(x_axis,y_axis, color="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.legend()
```

## Bubble Charts

```python
# Create bubble charts using OO interface
fig, ax = plt.subplots()
# size of bubble : s
# color of bubble: c, edgecolors, alpa to set transparancy

ax.scatter(y_axis, x_axis,s=[i*5 for i in y_axis], c='#87ceeb', alpha=0.8, edgecolors="k", linewidths=2.0, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare in Months")
ax.invert_yaxis()
ax.set_xlim(5,50)
ax.legend()
```

## Create Pie charts

```python
# Using MATLAB methods
plt.pie(y_axis,labels=x_axis)
```

```python
# Increase teh size of chart to see percentages properly
plt.subplots(figsize=(8, 8))
# To pop out the seventh month, July, we use a decimal-point value of 0.2.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#The length of the tuple should be equal to the number of wedges in the pie chart.
#The value specified to "explode" a wedge is the fraction of the radius from the center of the pie for each wedge.
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
#add the percentage of each wedge on the pie chart, autopct

```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create pie chart using OO interface
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
ax.pie(y_axis, colors=colors, explode=explode_values, shadow=True, startangle=90, counterclock=False, labels=x_axis, autopct="%.1f%%")
```
